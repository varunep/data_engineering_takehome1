```python
import findspark
findspark.init()
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder. \
    appName("pyspark-1"). \
    getOrCreate()
```

### Read data

```python
df = spark.read.csv("/dataset/nyc-jobs.csv", header=True, inferSchema=True)
df.printSchema()
```

```python
from pyspark.sql.functions import col, when, lit, year, month, to_date,split,datediff,current_date,row_number
from pyspark.sql.functions import col, trim, when, to_date, avg, max,min
from pyspark.sql.types import DoubleType
import matplotlib.pyplotas as plt
import seaborn as sns
import pandas as pd
from pyspark.sql import Window
```

### Functions to clean, wrangling,

```python
"""
Cleans the input DataFrame by removing nulls, trimming whitespaces, and filling missing values.
"""

def clean_data(df):   
    # Remove leading and trailing whitespace from all string columns
    string_columns = [c for c, t in df.dtypes if t == 'string']
    for col_name in string_columns:
        df = df.withColumn(col_name, trim(col(col_name)))
    
    #This will drop the records thathas fewer than threenot null records
    df = df.dropna(thresh=3)
        
    # 1. Fill missing values
    df = df.fillna({
        "Agency": "Unknown",
        "Job Category": "Unknown",
        "Recruitment Contact": "Unknown",
        "Salary Range From": 0,
        "Salary Range To": 0
    })
    return df

df_cleaned = clean_data(df)
df_cleaned.select("Agency","Job Category","Recruitment Contact","Salary Range From").show()
```

```python
"""
Wrangles the DataFrame by creating new features like average salary and normalized salary, and removing unnecessary columns.
"""

def wrangle_data(df):
    # 1. Create a Salary Range Feature: Average Salary
    df = df.withColumn("average_salary", (col("Salary Range From") + col("Salary Range To")) / 2)
    
    #2 Remove unnecessary columns that are not used in any KPI
    df = df.drop("Posting Type", "# Of Positions", "Title Code No", 
                 "Additional Information", "To Apply", "Recruitment Contact")


    # 3. Create a normalized salary column (per year if frequency is weekly or hourly)
    df = df.withColumn("Normalized_Salary", 
                       when(col("Salary Frequency") == "Annual", col("average_salary"))
                       .when(col("Salary Frequency") == "Hourly", col("average_salary") * 40 * 52)
                       .when(col("Salary Frequency") == "Weekly", col("average_salary") * 52)
                       .otherwise(col("Salary Range From")))
    
    # 4  Categorize salaries into different bands (e.g., low, medium, high) to help in analyzing trends or performing segmentation
    df = df.withColumn("salary_band",
                   when(col("normalized_salary") < 50000, "Low")
                   .when(col("normalized_salary").between(50000, 100000), "Medium")
                   .otherwise("High"))

    return df

df_wrangled = wrangle_data(df)
df_wrangled.select("Normalized_Salary","average_salary","salary_band").show()
```

```python
"""
Adds new features to the DataFrame, such as job level, work location breakdown, and days since posting, full time indicator.
"""

def feature_engineering(df):
    
    # 1. Extract Year and Month from Posting Date
    df = df.withColumn("Posting_Date", to_date(col("Posting Date"), "yyyy-MM-dd"))
    df = df.withColumn("Posting_Year", year(col("Posting_Date")))
    df = df.withColumn("Posting_Month", month(col("Posting_Date")))
    
    # 2. Flag for Full-Time/Part-Time Jobs
    df = df.withColumn("is_full_time", when(col("Full-Time/Part-Time indicator") == "F", 1).otherwise(0))
    
    
    # 3.  Convert  numerical value into job level information (e.g., Junior, Mid, Senior) 
    df = df.withColumn("job_level_numeric", 
                    when(col("Level") == 0, "Trainee")
                   .when(col("Level") == 1, "Junior")
                   .when(col("Level") == 2, "Mid")
                   .when(col("Level") == 3, "Senior")
                   .otherwise("Partner"))  # M1,M2,M3 for Highest management level
    
    #4.    Extract specific components from the "Work Location" column, such as city or state
    df = df.withColumn("work_city", split(col("Work Location"), ",")[0])
    df = df.withColumn("work_state", split(col("Work Location"), ",")[1])
    
    #5.   Create a feature that indicates how recently the job was posted. This can help in analyzing trends or changes in job market demand
    
    df = df.withColumn("days_since_posting",
                   datediff(current_date(), to_date(col("Posting Date"), "MM/dd/yyyy")))

    
    return df


df_featured = feature_engineering(df)
df_featured.select("job_level_numeric","is_full_time","Posting_Date","Posting_Year","work_city","work_state","days_since_posting").show()
```

```python
"""
Executes the full processing pipeline: cleaning, wrangling, and feature engineering.
"""

#combining all the outputs of the functions into a final_df
def process_pipeline(df):
    
    try:
        # Step 1: Clean data
        cleaned_df = clean_data(df)
        print("Data cleaning completed successfully.")

        # Step 2: wrangle data
        wrangle_df = wrangle_data(cleaned_df)
        print("Data wrangling completed successfully.")

        # Step 3: adding feature
        final_df = feature_engineering(wrangle_df)
        print("Feature engineering completed successfully.")

        return final_df
    
    except Exception as e:
        # Log the error message
        print(f"An error occurred: {e}")


print("done")
```

```python
# If the script is being run directly, process the data
if __name__ == "__main__":
    try:

        initial_df = spark.read.csv("/dataset/nyc-jobs.csv", header=True, inferSchema=True)
        # Process the DataFrame through the pipeline
        final_df = process_pipeline(initial_df)
        #write the data to destination
        final_df.write.mode("overwrite").option("header", "true").csv("/dataset/output.csv")

        print("Data written successfully")

        """KPI: number of jobs posting per category"""                  

        top_10_job_categories = final_df.groupBy("Job Category") \
            .count() \
            .orderBy(col("count").desc()) \
            .limit(10)

        top_10_job_categories.show()

        import matplotlib.pyplot as plt
        import seaborn as sns

        top_categories_pd = top_10_job_categories.toPandas()
        plt.figure(figsize=(40, 6))
        sns.barplot(x='count', y='Job Category', data=top_categories_pd, palette="viridis")
        plt.title('Top 10 Job Categories by Number of Postings')
        plt.xlabel('Number of Postings')
        plt.ylabel('Job Category')
        plt.show()


        """Salary Distribution per Job Category"""

        salary_distribution = final_df.groupBy("Job Category") \
            .agg(
                min("Salary Range From").alias("Min Salary"),
                max("Salary Range To").alias("Max Salary"),
                avg("average_salary").alias("Average Salary")
            )

        salary_distribution.show()

        salary_distribution_pd = salary_distribution.toPandas()

        # Plotting the salary distribution per job category
        plt.figure(figsize=(60, 10))
        sns.barplot(x='Job Category', y='Average Salary', data=salary_distribution_pd, palette="Blues_d")

        plt.title('Salary Distribution per Job Category')
        plt.xlabel('Job Category')
        plt.ylabel('Average Salary')
        plt.xticks(rotation=45, ha="right")
        plt.show()


        """job posting having the highest salary per agency"""

        # Create a column for the maximum salary for each job posting
        df = df.withColumn("max_salary", (col("Salary Range To")))

        # Define a window specification to partition by Agency and order by max_salary in descending order
        windowSpec = Window.partitionBy("Agency").orderBy(col("max_salary").desc())

        # Use row_number to assign a rank to each job posting within each agency
        df_with_rank = df.withColumn("rank", row_number().over(windowSpec))

        # Filter to keep only the highest salary job posting per agency (rank == 1)
        highest_salary_per_agency = df_with_rank.filter(col("rank") == 1) \
            .select("Agency", "Job ID", "Business Title", "max_salary")

        highest_salary_per_agency.show()

        # Test Case 1: Check if the highest salary per agency is correctly identified
        assert df_with_rank.filter(col("rank") == 1).count() == df.select("Agency").distinct().count(), "Test Case 1 Failed"

        # Test Case 2: Validate that no other ranks except 1 are present in the final DataFrame
        assert highest_salary_per_agency.filter(col("rank") != 1).count() == 0, "Test Case 2 Failed"

        print("All test cases passed!")

        #plotting for visualisation

        highest_salary_per_agency_pd = highest_salary_per_agency.toPandas()

        # Plotting the highest salary per agency
        plt.figure(figsize=(14, 8))
        sns.barplot(x='Agency', y='max_salary', data=highest_salary_per_agency_pd, palette="Greens_d")
        plt.title('Highest Salary per Agency')
        plt.xlabel('Agency')
        plt.ylabel('Highest Salary')
        plt.xticks(rotation=45, ha="right")
        plt.show()


        """Average Salary per Agency for the Last 2 Years """

        from pyspark.sql.functions import year

        df_filtered = final_df.filter(year(col("Posting Date")) >= year(current_date()) - 2)
        average_salary_per_agency = df_filtered.groupBy("Agency") \
            .agg(avg("average_salary").alias("Average Salary"))

        average_salary_per_agency.show()

        """ Highest Paid Skills in the US Market """

        from pyspark.sql.functions import explode, split,regexp_replace

        df_skills = final_df.withColumn(
            "SkillsArray",
            split(
                regexp_replace(col("Preferred Skills"), r"\d+\.\s*", "|"),  # Replace '1. ', '2. ' with '|'
                r'\s*\|\s*'  # Split by '|', handling optional spaces around '|'
            )
        )

        df_skills = df_skills.withColumn("Skill", explode(col("SkillsArray")))


        highest_paid_skills = df_skills.groupBy("Skill") \
            .agg(avg("average_salary").alias("Average Salary")) \
            .orderBy(col("Average Salary").desc()) \
            .limit(10)

        highest_paid_skills.select("Skill","Average Salary").show()

        highest_paid_skills_pd = highest_paid_skills.toPandas()

        # Truncating the Skill names to 20 characters as texts are too long
        highest_paid_skills_pd['Skill'] = highest_paid_skills_pd['Skill'].apply(lambda x: x[:20] + '...' if len(x) > 20 else x)

        # Plotting the highest paid skills
        plt.figure(figsize=(14, 8))
        sns.barplot(x='Average Salary', y='Skill', data=highest_paid_skills_pd, palette="Blues_d")

        plt.title('Top 10 Highest Paid Skills in the US Market')
        plt.xlabel('Average Salary ($)')
        plt.ylabel('Skill')

        plt.yticks(fontsize=10, rotation=30)
        plt.show()
                                    
#         spark.stop()

    except Exception as e:
        print(f"An error occurred: {e}")
        spark.stop()
        raise e
```
